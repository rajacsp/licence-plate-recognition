```python
edged = cv2.Canny(gray, 30, 200)
plt.imshow(edged, cmap = 'gray')
```

```python
cv2.imshow('windows1', edged)
cv2.imshow('windows2', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
